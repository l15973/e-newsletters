```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs, load_iris
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
X, y_true = make_blobs(n_samples=300, n_features=2, centers=4,
                       cluster_std=0.60, random_state=0, shuffle=True)
```

```python
plt.scatter(X[:,0], X[:,1], c="black", marker="o", s=40)
```

```python
estimator = KMeans(n_clusters=4, init="k-means++", n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_predict = estimator.fit_predict(X)
```

```python
y_predict
```

```python
plt.scatter(X[y_predict==0, 0], X[y_predict==0, 1], s=40, c = "lightgreen", marker = 's', label='cluster 1')
plt.scatter(X[y_predict==1, 0], X[y_predict==1, 1], s=40, c = "orange", marker = 'o', label='cluster 2')
plt.scatter(X[y_predict==2, 0], X[y_predict==2, 1], s=40, c = "lightblue", marker = 'v', label='cluster 3')
plt.scatter(X[y_predict==3, 0], X[y_predict==3, 1], s=40, c = "yellow", marker = 'p', label='cluster 4')
plt.scatter(estimator.cluster_centers_[:,0], estimator.cluster_centers_[:,1], s=40, c = "red", marker = '*', label='centroids')
plt.legend()
plt.grid()
```

```python
estimator.cluster_centers_
```

```python
estimator.inertia_
```

## Elbow Method

```python
all_sse = []
for clusterNum in range(1,11):
    estimator = KMeans(n_clusters=clusterNum, init="k-means++", n_init=10, max_iter=300, random_state=0)
    estimator.fit(X)
    all_sse.append(estimator.inertia_)
plt.plot(range(1,11), all_sse, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
```
